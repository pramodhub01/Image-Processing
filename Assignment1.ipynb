1. Read a color image.
2. Display it.
3. Resize it.
4. Convert image from Color to Gray scale.
4. Display original color image, gray image and resized image in a subplot.

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

1. Read a color image.

```python
#Reading a color image.....
img = cv2.imread(r"C:\Users\HP\Desktop\DIPLAB\Testimage\Lenna.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#Displaying the image.........
plt.figure(figsize=(8,3))
plt.imshow(img)
plt.axis('off')

print(img.size)
```

3. Resize it.

```python
img_resized = cv2.resize(img, (256,256))

plt.figure(figsize=(8,3))
plt.imshow(img_resized)
print(img_resized.shape)
print(img_resized.size)
```

```python
# Use the cvtColor() function to grayscale the image 
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

plt.figure(figsize=(8,3))
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
```

Display original color image, gray image and resized image in a subplot.

```python
plt.figure(figsize=(8,3))
plt.subplot(1,3,1)
plt.title("Input image")
plt.imshow(img)
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Gray image")
plt.imshow(gray_image, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Resized image")
plt.imshow(img_resized)
plt.axis('off')
```
